```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/house-prices.csv")
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
df[df.isna().any(axis=1)]
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df['Price'])
```

```python
df.columns
```

```python
X=df[['Home', 'Price', 'SqFt', 'Bedrooms', 'Bathrooms', 'Offers']]
```

```python
y=df['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pd.DataFrame(model.coef_,X.columns, columns=['Coeficient'])
```

```python
y_pred=model.predict(X_test)
```

```python
plt.scatter(y_test, y_pred)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
MAE=metrics.mean_absolute_error(y_test,y_pred)
MSE=metrics.mean_squared_error(y_test,y_pred)
RMSE=np.sqrt(MSE)
```

```python
pd.DataFrame([MAE,MSE,RMSE],['MAE','MSE','RMSE'],columns=['Metrics Value'])
```

```python
test_residuals=y_test-y_pred
```

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='r',linestyle='--')
```

```python
sns.distplot(test_residuals)
```
